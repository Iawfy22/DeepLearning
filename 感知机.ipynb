> 与回归的区别：

> 与softmax区别：
> softmax是多分类，输出概率，而感知机是二分类，输出1或0（是与不是）
#### 感知机的训练

通过$w <- w + y_ix_i$和$b <- b + y_i$来不断更新参数，直至分类正确<br>
等价于使用批量大小为1的梯度下降，并使用以下损失函数：
$$ l(y, x, w) = max(0, -y<w, x>)$$

#### 收敛定理
判断什么时候停止，会不会停止的问题<br>
基本假设：
- 数据在半径r内
- 余量$\rho$分类为两类，且所有分类都要正确，即$t(x^Tw + b) >= \rho$
- 感知机保证在$\frac{r^2+1}{\rho ^2}$步后收敛

#### 感知机的问题
感知机不能拟合XOR问题，即异或问题，它只能产生线性分割面，即上图分割猫狗的的分割线必须是直线，不能是曲线（对应二维的问题，分割面必须是直线）
